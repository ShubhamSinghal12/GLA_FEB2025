````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
X = np.linspace(-1,1,10)
Y1 = 2*X-3
Y2 = 4*(X**2)-5
Y3 = X**3 - X**2 + X - 5
````

````python
plt.plot(X,Y1,'g--o',label = '1')
plt.plot(X,Y2,'b-.^',label = '2')
plt.plot(X,Y3,'y-d',label = '3')
plt.xlabel('Length')
plt.ylabel('Width')
# plt.xlim(0,11)
plt.grid()
plt.legend(loc = 'upper left')
# plt.show()
````

````python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
x = np.arange(1,11)
y1 = 3*x+5
y2 = 2*x+3
plt.plot(x,y1,label = 'Price',color = 'r',marker = '*',markersize = 15,markerfacecolor = 'b')
plt.plot(x,y2,label = 'Line',color = 'g',marker = 'o',linestyle=':')

plt.xlabel('X',fontdict=font)
plt.ylabel('f(X) = Y',fontdict = font)
plt.title("Line Plot",fontdict = font)
plt.legend()
plt.grid()
# plt.savefig('LinePlot.jpg')
plt.show()
````

````python
X = np.linspace(-1,1,100)
sin = np.sin(2*np.pi*X)
cos = np.cos(2*np.pi*X)
````

````python
plt.figure(figsize=(10,4))
plt.subplot(2,1,1)
plt.plot(X,sin)
plt.grid()
plt.title('Sin Wave')

plt.subplot(2,1,2)
plt.plot(X,cos)
plt.grid()
plt.title('Cos Wave')

plt.subplots_adjust(hspace=0.5)
plt.show()
````

````python
X = np.linspace(-1,1,10)
Y1 = 2*X-3
plt.scatter(X,Y1)
plt.show()
````

````python
labels = ['Silver','Gold','Platinum']
price1 = [100,80,120]
price2 = [90,95,130]
X = np.array([1,2.5,4])

plt.bar(X-0.25,price1,width=0.5)
plt.bar(X+0.25,price2,width=0.5)
plt.ylabel('Price')
plt.xticks(X,labels)
plt.show()
````

````python
plt.pie([10,20,10,30,20],explode=[0,0,0,0.1,0],shadow=True,labels=['DBMS','DSA','CSA','SED','SET'],autopct='%.2f%%')
plt.show()
````

````python
tips = sns.load_dataset('tips')
````

````python
sns.barplot(tips,x='day',y='total_bill',hue = 'sex',estimator='mean')
plt.show()
````
